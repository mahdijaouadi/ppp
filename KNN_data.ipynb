```python
def plot_chart(stock):
    fig = go.Figure()
    # Add candlestick data
    fig.add_trace(go.Candlestick(
        x=stock['Datetime'],
        open=stock['Open'],
        high=stock['High'],
        low=stock['Low'],
        close=stock['Close']
    ))

    fig.show()

def get_file_name_if_size_greater_than(file_path, size_threshold_kb):
    try:
        # Get the size of the file in bytes
        size_in_bytes = os.path.getsize(file_path)

        # Convert the size to kilobytes
        size_in_kb = size_in_bytes / 1024.0

        # Check if the file size is greater than the threshold
        if size_in_kb > size_threshold_kb:
            file_name_without_extension = os.path.splitext(os.path.basename(file_path))[0]
            return file_name_without_extension

    except FileNotFoundError:
        return None  
def get_stock_data(ticker):
    stock=pd.read_csv(f'C:\\Users\\mahdi\\Desktop\\stock data\\Data2023_1D\\{ticker}.csv')
    stock.drop(['Unnamed: 0'],axis=1,inplace=True)
    return stock

first_plane0=np.load('first_plane0.npy')
second_plane0=np.load('second_plane0.npy')
third_plane0=np.load('third_plane0.npy')

first_plane1=np.load('first_plane1.npy')
second_plane1=np.load('second_plane1.npy')
third_plane1=np.load('third_plane1.npy')

first_plane2=np.load('first_plane2.npy')
second_plane2=np.load('second_plane2.npy')
third_plane2=np.load('third_plane2.npy')
```

```python
coef=np.ones(11)
coef[10]=3
# coef[28]=1
# coef[27]=1
# coef[26]=1
# coef[25]=1
def distance(x1,x2,coef):
    d=math.sqrt(sum((coef[i]*(x1[i]-x2[i])**2) for i in range(11)))
    return d
def cosine(x1,x2):
    cos=np.dot(x1.T,x2)/(np.linalg.norm(x1)*np.linalg.norm(x2))
    return cos
```
